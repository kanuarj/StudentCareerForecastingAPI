```python
import pandas as pd
df = pd.read_csv('./../Datasets/data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.Location.unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
locs = le.fit_transform(df[['Location']])
```

```python
df['Encoded_Location'] = locs
```

```python
df
```

```python
df[['Andheri', 'Bandra', 'Borivali', 'Dadar', 'Vashi']] = pd.get_dummies(df.Encoded_Location)
```

```python
df
```

```python
df.rename(columns={'CS-CAP1':'Marks'}, inplace=True)
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
X_data = df[['Marks', 'Andheri', 'Bandra', 'Borivali', 'Dadar', 'Vashi']]
y_data = df['Institute Name']
```

```python
model.fit(X_data, y_data)
```

```python
model.score(X_data, y_data)
```

```python
model.predict([[23,0,0,0,1,0]])
```
